```python
import pandas as pd
from bs4 import BeautifulSoup

import requests as reg
import re
```

```python
html = reg.get('http://smspalsudanpenipu.blogspot.com/p/daftar-nomor-sms.html')
```

```python
s = BeautifulSoup(html.text, 'lxml')
```

```python
isi = s.body.find('div', attrs={'class':'post-outer'}).text
```

```python
f_hp = re.compile(r'\d\d\d\d\d\d\d\d\d\d\d\d')
f_nama = re.compile(r':\s(\w+\s\w*\s\w*\s)')
```

```python
nomorhp = f_hp.findall(isi)
namaP = f_nama.findall(isi)
```

```python
# print('nomor hp penipuan \n', *nomorhp, sep="\n")
print('nama',*namaP, sep='\n')
```

```python
print('nomor hp penipuan \n', *nomorhp, sep="\n")
```
